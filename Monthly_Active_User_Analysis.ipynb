```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_excel('/content/Active Users (4) (1) (1).xls')
```

```python
df.head(1760)
```

MAU(t) – MAU(t – 1 month) = new(t) + resurrected(t) – churned(t)

```python
retained=np.zeros(len(df.columns),dtype=int)
churned=np.zeros(len(df.columns),dtype=int)
new=np.zeros(len(df.columns),dtype=int)
resurrected=np.zeros(len(df.columns),dtype=int)

for j in range(len(df.columns)):
  for i in df.index:
    if j==0:
      if pd.isna(df.iloc[i, j])==False:
        new[j]+=1
    else:
      if pd.isna(df.iloc[i, j])==False and pd.isna(df.iloc[i, j-1])==False:
        retained[j]+=1
      elif pd.isna(df.iloc[i, j]) and pd.isna(df.iloc[i, j-1])==False:
        churned[j]-=1
      elif pd.isna(df.iloc[i, j])==False and pd.isna(df.iloc[i,j-1]):
        for x in range(j):
          if pd.isna(df.iloc[i, x])==False:
            set=1
            break
        if set==0:
          new[j]+=1
        else:
          resurrected[j]+=1
        set=0
```

```python
qr=-100*((new+resurrected)/churned)
```

```python
myinfo=np.vstack([churned,new,resurrected,qr])
mycols=df.columns
myindex=['churned','new','resurrected','quickratio']
mydf=pd.DataFrame(data=myinfo,columns=mycols,index=myindex)
mydf.head()
```

```python
plt.bar(range(1,57), churned)
plt.bar(range(1,57), new)
plt.bar(range(1,57), resurrected)
plt.plot(range(1,57), qr)
plt.legend(['qr','churned','new','resurrected'])
plt.xlabel('weeks')
plt.ylabel('number of users')
plt.title('Active Users Chart')
plt.show()
```
